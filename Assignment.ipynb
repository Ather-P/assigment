```python
import os
import pandas as pd
import numpy as np
os.chdir('/content/drive/MyDrive/Eccentric Assignment')
```

```python
data = pd.read_csv('assignment.csv')
```

```python
from imblearn import over_sampling
```

```python
data = data.drop(['ID','Pincode','Exchange','Finance','Terms & Condition'],axis=1)
data['Created At'] = pd.to_datetime(data['Created At'])
data['Updated At'] = pd.to_datetime(data['Updated At'])
```

```python
data['Book Now'].value_counts()
X = data.drop('Book Now',axis=1)
Y = data['Book Now']
```

```python
data.head()
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=0)
X_resampled,Y_resampled = ros.fit_resample(X,Y)
```

```python
df = pd.concat([X_resampled,Y_resampled],axis =1)
```

```python
data['Variant ID'].unique()
```

```python
df.head()
```

```python
from pycaret.classification import *
s = setup(data = df, target = 'Book Now', session_id=123)
```

```python
best = compare_models()
```

```python
et = create_model('et')
```

```python
tuned_model = tune_model(et)
```

```python
evaluate_model(tuned_model)
```

```python
import pickle

Model = 'Tress_classifer_Booking.pkl'
pickle.dump(tuned_model,open(Model,'wb'))
```

```python
final_model = finalize_model(tuned_model)
save_model(final_model,"Prediction_model")
```
